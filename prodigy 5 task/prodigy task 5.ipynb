```python
# Visualize accident count by month
monthly_accidents = data.drop(['STATE/UT', 'YEAR', 'TOTAL'], axis=1)
monthly_accidents_sum = monthly_accidents.sum()

plt.figure(figsize=(10, 6))
monthly_accidents_sum.plot(kind='bar', color='skyblue')
plt.title('Total Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Total Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize accident count by state/UT
accidents_by_state = data.groupby('STATE/UT')['TOTAL'].sum().sort_values(ascending=False)[:10]

plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_by_state.values, y=accidents_by_state.index, palette='coolwarm')
plt.title('Top 10 States/UTs with Highest Total Accidents')
plt.xlabel('Total Accidents')
plt.ylabel('State/UT')
plt.show()
```

```python
# Visualize accidents by year
accidents_by_year = data.groupby('YEAR')['TOTAL'].sum()

plt.figure(figsize=(10, 6))
accidents_by_year.plot(kind='line', marker='o', color='green')
plt.title('Total Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Total Accidents')
plt.xticks(accidents_by_year.index)
plt.grid(True)
plt.show()
```

```python
# Extracting month-wise data for accidents by year
accidents_by_year = data.groupby('YEAR')[['JANUARY', 'FEBRUARY', 'MARCH', 'APRIL', 'MAY', 'JUNE', 'JULY', 'AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER']].sum()

# Reshaping the DataFrame for plotting
accidents_by_year = accidents_by_year.transpose()

plt.figure(figsize=(18, 12))
accidents_by_year.plot(marker='o', color='blue')
plt.title('Total Accidents Over Time')
plt.xlabel('Month')
plt.ylabel('Total Accidents')
plt.xticks(rotation=45)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_columns = data.select_dtypes(include='number')

# Calculate correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
